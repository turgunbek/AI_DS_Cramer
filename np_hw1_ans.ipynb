````python
n = 5
np_arr_7 = np.eye(n)
print(np_arr_7)
````

````python
#   :)
def get_eye(n):
    eye = np.zeros((n, n))
    for i in range(n):
        eye[i, i] = 1
    return eye
````

````python
print(get_eye(5))
````

````python
def get_diag_n2(n):
    res = np.arange(1, n**2+1).reshape((n,n))
    return np.diag(np.diag(res))
#      :
#     return res*np.eye(n)
````

# Task_9**

```
n = 4

array([[ 1,  2,  3,  4],
       [ 0,  6,  7,  8],
       [ 0,  0, 11, 12],
       [ 0,  0,  0, 16]])
```



````python
def get_upper_tri_n2(n):
    res = np.arange(1, n**2+1).reshape((n,n))
    return np.triu(res)
````

````python
print(get_upper_tri_n2(4))
````
