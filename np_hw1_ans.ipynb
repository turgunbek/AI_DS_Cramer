# Task_7

Придумайте алгоритм создания единичной матрицы произвольного размера. Можете оформить задачу в виде функции. За основу можно взять нулевую матрицу и использовать инструменты предыдущего таска. Алгоритм должен одинаково работать для квадратных матриц любой размерности.

Единичная матрица - это КВАДРАТНАЯ матрица, которая имеет главную диагональ из единиц.

image.png

````python
n = 5
np_arr_7 = np.eye(n)
print(np_arr_7)
````

````python
# если придумывать велисипед:)
def get_eye(n):
    eye = np.zeros((n, n))
    for i in range(n):
        eye[i, i] = 1
    return eye
````

````python
print(get_eye(5))
````

# Task_8*

Придумайте алгоритм создания квадратной матрицы произвольного размера с числами на множестве $[1, n^2]$ у которой все элементы кроме главной диагонали заменены нулем. Можете оформить задачу в виде функции. За основу можно взять  матрицу нулей и использовать инструменты предыдущего таска. Алгоритм должен одинаково работать для квадратных матриц любой размерности.

```
n = 5

array([[ 1,  0,  0,  0,  0],
       [ 0,  7,  0,  0,  0],
       [ 0,  0, 13,  0,  0],
       [ 0,  0,  0, 19,  0],
       [ 0,  0,  0,  0, 25]])
```


*ВАЖНО ИМЕТЬ В ВИДУ*

Если вы с помощью слайсинга выделили фрагмент матрицы, вы можете сделать замену выбранных элементов на какое либо число. Но вы также можете выполнить замену, подав список или массив значений:



image.png

````python
def get_diag_n2(n):
    res = np.arange(1, n**2+1).reshape((n,n))
    return np.diag(np.diag(res))
#     или так:
#     return res*np.eye(n)
````

````python
get_diag_n2(5)
````

# Task_9**

Придумайте алгоритм создания квадратной матрицы произвольного размера с числами на множестве $[1, n^2]$ у которой все элементы под главной диагональю заменены нулем. Можете оформить задачу в виде функции. За основу можно взять  матрицу с незануленными элементами и использовать инструменты предыдущего таска. Алгоритм должен одинаково работать для квадратных матриц любой размерности.



```
n = 4

array([[ 1,  2,  3,  4],
       [ 0,  6,  7,  8],
       [ 0,  0, 11, 12],
       [ 0,  0,  0, 16]])
```



````python
def get_upper_tri_n2(n):
    res = np.arange(1, n**2+1).reshape((n,n))
    return np.triu(res)
````

````python
print(get_upper_tri_n2(4))
````

# Task_10

Из предложенного массива отберите числа по условию:
* $x_i \geq 50$.
* $25 < x_i \leq 75$
* $x\in (0, 20] \cup [80, 100]$

````python
arr = np.random.randint(1, 100, 50)
arr
````

````python
np_arr_10_1 = arr[arr>=50]
np_arr_10_2 = arr[(25<arr) & (arr<=75)]
np_arr_10_3 = arr[(0<arr) & (arr<=20) | (80<=arr) & (arr<=100)]
print(np_arr_10_1)
print(np_arr_10_2)
print(np_arr_10_3)
````
